```python
import pandas as pd
import json
```

```python
odds = pd.read_csv('/content/drive/MyDrive/MC536 - Grupo GOLDB/Tabelas/BRA.csv')
```

```python
odds = odds.drop(['Country', 'League', 'Date', 'Time'], 1)
corrigir_ids = {'Botafogo RJ':'Botafogo', 'Fortaleza':'Fortaleza EC', 'Chapecoense-SC':'Chapecoense-sc', 'Flamengo RJ':'Flamengo', 'Athletico-PR':'Atletico Paranaense', 'Vasco':'Vasco DA Gama'}
```

```python
odds = odds.drop(odds[odds.Season != 2019].index)
odds = odds.replace(corrigir_ids)
odds = odds.drop(['Season'], 1)
odds['PartidaID'] = ""

odds
```

```python
partidas_json_file = open('/content/drive/MyDrive/MC536 - Grupo GOLDB/JSON/fixtures.json')
partidas_json = json.load(partidas_json_file)
fixtures = partidas_json['response']
partidas = dict()
for fixture in fixtures:
  times_str = str(fixture['teams']['home']['name']) + " X "+ str(fixture['teams']['away']['name'])
  partidas[times_str.upper()] = str(fixture['fixture']['id'])
for index, row in odds.iterrows():
    times_key = str(row['Home']).upper() + " X "+ str(row['Away']).upper()
    odds.at[index, 'PartidaID'] = partidas[times_key]
```
